```python
import sys, os
sys.path.append(os.path.dirname(os.getcwd()))
from data.nela2017 import Nela2017
from utils.text.preprocess import preprocess
from utils.plotting.plotting import wordcloud
```

```python
ROOT = r"C:\Users\alexander.paskal\datasets\NELA2017"
dataset = Nela2017(ROOT)
```

```python
def org_titles(org):
    dataset.organizations(org)
    titles = [article.title for article in dataset]
    text = " ".join(titles)
    text = preprocess(text)
    return text

fox_titles = org_titles("FOX NEWS")
cnn_titles = org_titles("CNN")
```

```python
wordcloud(" ".join(fox_titles))
```

```python
wordcloud(" ".join(cnn_titles))
```

```python
# takes about 20 seconds to run
def org_text(org):
    dataset.organizations(org)
    text = [article.text for article in dataset]
    text = " ".join(text)
    text = preprocess(text)
    if isinstance(text, list):
        text = " ".join(text)
    return text

import datetime
dataset.dates = (datetime.datetime(2017, 4, 1), datetime.datetime(2017, 4, 3))
fox_text = org_text("FOX NEWS")
cnn_text = org_text("CNN")
```
